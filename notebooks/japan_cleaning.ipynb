### 1. Load and inspect raw data

```python
import pandas as pd
data1 = pd.read_excel("../data/000280216.xlsx",sheet_name = '認証品目リスト',header =1 )
print(data1.head())
```

```python
# Dataset shape
print("Dataset shape:")
print(data1.shape)

# Column names
print("\nColumn names:")
print(data1.columns.tolist())

# Data types
dtype_df = pd.DataFrame({"column_name": data1.columns, "data_type": data1.dtypes.values})
dtype_df.style.hide(axis="index")
```

### 2. Standardize column names and add the data source

```python
data1.columns
```

```python
column_mapping = {
    '認証機関コード': 'certification_body_code',
    '認証番号': 'certification_number',
    '認証年月日': 'certification_date',
    '販売名': 'sales_name',
    '一般的名称': 'general_name',
    '業者名_認証取得者': 'vendor_name_certified_person',
    '法人番号': 'corporate_number',
    '業者名_選任外国製造医療機器等製造販売業者': 'company_name_designated_foreign_manufactured_medical_device_manufacturer_and_distributor',
    '法人番号.1': 'corporate_number_2',
    '承認からの移行認証': 'transition_from_authorization_to_certification',
    '承継品目': 'succession_item',
    '承継年月日': 'date_of_succession',
    '承継時認証機関変更': 'change_of_certification_body_at_time_of_succession',
    '認証整理日': 'certification_summary_date',
    '認証取消日': 'certification_revocation_date'
}
```

```python
data1 = data1.rename(columns=column_mapping)
print(data1.columns)
```

```python
data1["source_file"] = "000280216.xlsx"
data1["source_url"] = "https://www.pmda.go.jp/review-services/drug-reviews/about-reviews/devices/0026.html?"
```

### 3. Review missing values and preserve business-meaningful nulls

```python
data1.isnull().sum()
```

### 4. Standardize date columns

```python
date_columns = [
    "certification_date",
    "date_of_succession",
    "certification_summary_date",
    "certification_revocation_date"
]
for col in date_columns:
    data1[col] = pd.to_datetime(
        data1[col],
        errors="coerce"
    )
data1.dtypes
```

### 5.Deduplication
#### a. Define duplicate key columns
#### b. Flag duplicate groups using `duplicate_flag`
#### c. Review flagged duplicate records  
#### d. Remove confirmed duplicates while keeping the first occurrence  
#### e. Recheck the dataset for remaining duplicates

```python
#a. Define duplicate key columns
duplicate_keys = ["certification_number", "sales_name", "vendor_name_certified_person"]
data1.duplicated(subset=duplicate_keys, keep= False).sum()
```

```python
#b. Flag duplicate groups using `duplicate_flag`
data1['duplicate_flag'] = data1.duplicated(subset=duplicate_keys, keep=False)
```

```python
#c. Review flagged duplicate records 
duplicates = data1[data1['duplicate_flag'] == True]
duplicates.sort_values(duplicate_keys).head(5)
```

```python
#d. Remove confirmed duplicates while keeping the first occurrence  
cleaned_japan = data1.drop_duplicates(subset=duplicate_keys, keep='first')
```

```python
#e. Recheck the dataset for remaining duplicates
print('before:',len(data1))
print('after:',len(cleaned_japan))
print('remove:',len(data1)-len(cleaned_japan))
```

```python
cleaned_japan[cleaned_japan['duplicate_flag'] == True].head()

cleaned_japan.duplicated(subset = duplicate_keys, keep=False).sum()
```

```python
#Export dataset to excel
cleaned_japan.head()
cleaned_japan.to_excel('../data/cleaned_japan.xlsx',index=False)
```

### 6. Export cleaned dataset to SQLite

```python
import sqlite3
conn=sqlite3.connect('../database/Japan_medical_devices.sqlite')
cleaned_japan.to_sql('devices',conn,if_exists='replace',index=False)
conn.close()
```
